```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def MC(m,n,k):
  nt=0
  nr=0
  na=0
  mu0=1
  x0=0
  npart=10**(k+1)
  h=10**m
  s=1
  sc=10**(-n)
  distrib=np.zeros(100)


  for i in range(npart):
    end=0
    x=x0
    mu=mu0
    
    while(end==0):
      lambd=-1/s*np.log(np.random.rand())
      x=x+lambd*mu
      if (x>h):
        nt=nt+1
        end=1
      if (x<0):
        nr+=1
        end=1
      if ((x<=h)and(x>=0)):
        if (np.random.randn()<(sc/s)):
          end=1
          na=na+1
      mu=2*np.random.rand()-1
    

    if ((x<=h)and(x>=0)):
      distrib[int(x/h*100)]+=1
  return nt/npart,nr/npart,na/npart,distrib
```

```python
_,_,_,distr=MC(1,1,3)
```

```python
distr
```

```python
fig, axs = plt.subplots(2, 2)
fig.set_figheight(12)
fig.set_figwidth(32)

pt,pr,pa,distr=MC(1,2,3)
axs[0, 0].plot(np.linspace(0,10**(1),100),distr/(10**(4)*pa))
axs[0,0].set_title(1)
print("График номер 1 m=1,n=2,k=3   pt="+str(pt)+"  pr="+str(pr)+"  pa="+str(pa))

pt,pr,pa,distr=MC(0.05,1,3)
axs[0, 1].plot(np.linspace(0,10**(1),100),distr/(10**(4)*pa))
axs[0,1].set_title(2)
print("График номер 2 m=0.05,n=1,k=3   pt="+str(pt)+"  pr="+str(pr)+"  pa="+str(pa))

pt,pr,pa,distr=MC(0.5,1,3)
axs[1, 0].plot(np.linspace(0,10**(1),100),distr/(10**(4)*pa))
axs[1,0].set_title(3)
print("График номер 3 m=0.5,n=1,k=3   pt="+str(pt)+"  pr="+str(pr)+"  pa="+str(pa))

pt,pr,pa,distr=MC(1,0.01,3)
axs[1, 1].plot(np.linspace(0,10**(1),100),distr/(10**(4)*pa))
axs[1,1].set_title(4)
print("График номер 4 m=1,n=0.01,k=3   pt="+str(pt)+"  pr="+str(pr)+"  pa="+str(pa))
```
